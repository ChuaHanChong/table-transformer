```python
import json
from collections import Counter
from PIL import Image, ImageDraw
```

```python
img_path = "../data/PubTables-1M-Samples/PMC493271_table_0.jpg"
cell_ann_path = "../data/PubTables-1M-Samples/PMC493271_table_0.json"
word_ann_path = "../data/PubTables-1M-Samples/PMC493271_table_0_words.json"


colors = {
    "table": "brown",
    "table column": "red",
    "table row": "blue",
    "table column header": "magenta",
    "table projected row header": "cyan",
    "table spanning cell": "green",
    }


with open(cell_ann_path, "r") as f:
    cell_anns = json.load(f)

with open(word_ann_path, "r") as f:
    word_anns = json.load(f)

img = Image.open(img_path).convert("RGB")
img
```

```python
img_1 = img.copy()
img_2 = img.copy()
img_3 = img.copy()
img_4 = img.copy()
img_5 = img.copy()
img_6 = img.copy()

draw_1 = ImageDraw.Draw(img_1)
draw_2 = ImageDraw.Draw(img_2)
draw_3 = ImageDraw.Draw(img_3)
draw_4 = ImageDraw.Draw(img_4)
draw_5 = ImageDraw.Draw(img_5)
draw_6 = ImageDraw.Draw(img_6)

for ann in cell_anns["annotation"]["object"]:
    name = ann["name"]
    bbox = [float(v) for v in ann["bndbox"].values()]
    if name == "table":
        draw_1.rectangle(bbox, outline=colors[name])
    elif name == "table column":
        draw_2.rectangle(bbox, outline=colors[name])
    elif name == "table row":
        draw_3.rectangle(bbox, outline=colors[name])
    elif name == "table column header":
        draw_4.rectangle(bbox, outline=colors[name])
    elif name == "table projected row header":
        draw_5.rectangle(bbox, outline=colors[name])
    elif name == "table spanning cell":
        draw_6.rectangle(bbox, outline=colors[name])
    else:
        raise ValueError

display(img_1)
display(img_2)
display(img_3)
display(img_4)
display(img_5)
display(img_6)
```

```python
img_7 = img.copy()
draw_7 = ImageDraw.Draw(img_7)

for ann in word_anns:
    bbox = ann["bbox"]
    draw_7.rectangle(bbox, outline="red")

img_7
```
